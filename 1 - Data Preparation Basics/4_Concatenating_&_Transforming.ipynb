```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

Concatenating Data


```python
#Creates a 6x6 pandas DataFrame named df_obj with values from 0 to 35 and displays it.
df_obj = DataFrame(np.arange(36).reshape(6,6))
df_obj
```

```python
df_sec = DataFrame(np.arange(15).reshape(5,3))
df_sec
```

```python
#Concatenates df_obj and df_sec column-wise.
pd.concat([df_obj, df_sec], axis=1)
```

```python
pd.concat([df_obj, df_sec], axis=0)
```

Transforming Data


```python
df_obj.drop([0,2])
```

```python
df_obj.drop([0,2], axis=1)
```

Adding Data


```python
#Creates a pandas Series named series_ob with values from 0 to 5, sets its name to "added_vairable", and displays it.
series_ob = Series(np.arange(6))
series_ob.name = "added_vairable"
series_ob
```

```python
#Joins the pandas Series series_ob to the pandas DataFrame df_obj and displays the result.
var_added = DataFrame.join(df_obj, series_ob)
var_added
```

```python
#Concatenates var_added with itself, preserving the original index.
added_datatable = pd.concat([ var_added, var_added],ignore_index=False)
added_datatable
```

```python
#Concatenates var_added with itself, resetting the index.
added_datatable = pd.concat([ var_added, var_added],ignore_index=True)
added_datatable
```

Sorting Data

```python
#Sorts the rows of the DataFrame df_obj based on the values in column 5 in descending order.
df_sorted = df_obj.sort_values(by=(5), ascending=[False])
df_sorted
```
