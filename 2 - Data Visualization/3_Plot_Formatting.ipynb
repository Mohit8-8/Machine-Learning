```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
#ensures that plots created with Matplotlib are displayed directly within the output of the code cell
```

```python
rcParams['figure.figsize'] = 5 , 4
#subsequent plots created with Matplotlib will have a width of 5 inches and a height of 4 inches by default.
```

Defining axes, ticks and grids

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.plot(x,y)
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

# set the limits for the x and y axes
ax.set_xlim([1,9])
ax.set_ylim([0,5])

#set the positions of the tick marks on the x and y axes.
ax.set_xticks([1,2,3,4,5,6,7,8,9])
ax.set_yticks([0,1,2,3,4,5])

ax.grid()
ax.plot(x,y)
```

Generating multiple plots in one figure with subplots

```python
fig = plt.figure()

fig, (ax1, ax2) = plt.subplots(1,2)  #specifies the layout of the subplots. It means there will be 1 row and 2 columns of subplots.
ax1.plot(x)
ax2.plot(x,y)
```

Defining plot color

```python
x = range(1, 10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x, y)
```

```python
wide = [0.5, 0.5, 0.5, 0.9, 0.9, 0.9, 0.5, 0.5, 0.5]
color = ['salmon']
plt.bar(x, y, width=wide, color=color, align='center')
```

```python
z = [1,2,3,4,0.5]
plt.pie(z)
plt.show()
```

Customizing line styles

```python
x1 = range(1,10)
y1 = [9, 8, 7, 6, 5, 4, 3, 2, 1]

plt.plot(x, y)
plt.plot(x1,y1)
```

```python
plt.plot(x, y, ls = 'solid', lw=5)
plt.plot(x1,y1, ls='--', lw=10)
```
