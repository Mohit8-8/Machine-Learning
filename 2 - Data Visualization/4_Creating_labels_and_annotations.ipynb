```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 12,4
sns.set_style('whitegrid')
```

# Labeling plot features

The functional method

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x,y)

plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1,2,3,4,0.5]
veh_type = ['bicycle', 'motorbike','car','van','stroller']
plt.pie(z, labels=veh_type)
plt.show()
```

The object-oriented method

```python
address = '/content/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation = 60, fontsize = 'medium')
ax.set_title('miles per gallon of cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles per gallon')
```

# Adding a legend to your plot

The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc = 'best')
plt.show()

#A legend in a plot is a key that explains what the different colors, lines, or markers in the plot represent.
#It helps you understand what you're seeing in the visualization.
```

The object oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation = 60, fontsize = 'medium')
ax.set_title('miles per gallon of cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles per gallon')

ax.legend(loc = 'best')
```

Annotating the plot

```python
mpg.max()

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.set_ylim([0,45]) #he limits of the y-axis on the plot represented by the ax object, ranging from 0 to 45.

#adds an annotation to the plot at the point (19, 33.9) with the text 'Toyota Corolla', placing the text at (21, 35) and drawing a black arrow between the text and the data point.
ax.annotate('Toyota Corolla', xy=(19,33.9), xytext=(21,35),arrowprops=dict(facecolor='black', shrink=0.05))

ax.legend(loc='best')

```
