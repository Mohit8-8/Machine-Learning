```python
import numpy as np
import pandas as pd
```

```python
address = '/content/sample_data/mtcars.csv'
```

```python
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.index = cars.car_names

cars.head(15)
```

```python
carb = cars.carb
carb.value_counts()
```

```python
cars_cat = cars[['cyl', 'vs', 'am', 'gear', 'carb']]
#creates a new pandas DataFrame called cars_cat containing only the columns 'cyl', 'vs', 'am', 'gear', and 'carb' from the original cars DataFrame.
cars_cat.head()
```

```python
#groups the cars_cat DataFrame by the values in the 'gear' column and
#then calculates descriptive statistics for each group.
gears_group = cars_cat.groupby('gear')
gears_group.describe()
```

### Transforming variables to categorical data type

```python
cars['group'] = pd.Series(cars.gear, dtype='category')
#creates a new column in the cars DataFrame called 'group' and
#populates it with the data from the 'gear' column, converting the data type to 'category'.
```

```python
cars['group'].dtypes
```

```python
cars['group'].value_counts()
```

### Describing categorical data with crosstabs

```python
pd.crosstab(cars['am'], cars['gear'])
#This code generates a contingency table (crosstab)
# that shows the frequency distribution of the 'am' and 'gear' columns in the 'cars' DataFrame.
```
