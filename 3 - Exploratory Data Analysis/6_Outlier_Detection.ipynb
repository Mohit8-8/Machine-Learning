## Extreme value analysis using univariate methods

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

```python
address = '/content/sample_data/iris.data.csv'
df = pd.read_csv(filepath_or_buffer=address, header=None, sep=',')
#loads the Iris dataset, names the columns, separates the features and target variable, and displays the first few rows of the data.
df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'species']
x = df.iloc[:,0:4].values
y = df.iloc[:,4].values
df[:5]
```

Identifying outliers from tukey boxplots

```python
df.boxplot(return_type='dict')
plt.plot()
```

```python
sepal_width = x[:,1]
iris_outliers = (sepal_width>4)
df[iris_outliers]
```

```python
sepal_width = x[:,1]
iris_outliers = (sepal_width<2.05)
df[iris_outliers]
```

applying tukey outlier labelling

```python
pd.options.display.float_format = '{:.1f}'.format #formats all floating-point numbers to one decimal place when they are printed
x_df = pd.DataFrame(x)
print(x_df.describe())#calculates and prints descriptive statistics for each column in the x_df DataFrame
```
