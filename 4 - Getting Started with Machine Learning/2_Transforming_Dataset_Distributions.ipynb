```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, scale
```

```python
address = '/content/sample_data/mtcars.csv'

dataset = pd.read_csv(address)
dataset.head()
```

##Normalization
This is used to scale numerical data to a common range, typically between 0 and 1. This is done to ensure that no single feature dominates the learning process in machine learning algorithms due to its larger scale.

```python
plt.plot(dataset[['mpg']])
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(dataset[['mpg']])
```

```python
scaled_data = min_max_scaler.transform(dataset[['mpg']])
plt.plot(scaled_data)
```

##Standardization

```python
standard_scaler = scale(dataset[['mpg']])
plt.plot(standard_scaler)
```
